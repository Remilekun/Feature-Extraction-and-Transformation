```python
#Treating outliers

df_year3Med
```

```python
#Removing the feature Y

subData3 = df_year3Med.iloc[:, 0:64]
subData3
```

```python
Q1 = subData3.quantile(0.25)
Q3 = subData3.quantile(0.75)

# calculate the IQR
IQR = Q3 - Q1
IQR

lbound = Q1 - 1.5 * IQR
ubound = Q3 + 1.5 * IQR

def outlier_limits(col):
    Q3, Q1 = np.nanpercentile(col, [75, 25])
    IQR = Q3 -Q1
    ubound = Q3 + 1.5*IQR
    lbound = Q1 - 1.5*IQR
    return ubound, lbound
```

```python
# convert outlier values to missing values

for column in subData3.columns:
    if subData3[column].dtype !='object':
        ubound, lbound = outlier_limits(subData3[column])
        subData3[column] = np.where((subData3[column] > ubound) | (subData3[column] < lbound), np.nan, subData3[column])
```

```python
# check for missing values

subData3.isnull().sum()
```

```python
# Replace missing values removed from outliers with median

dfNew_year3 = subData3.fillna(subData3.median())
```

```python
# check if no missing value

dfNew_year3.isnull().sum()
```
